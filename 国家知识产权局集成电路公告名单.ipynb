```python
web = dict()
for n in [""] + list(range(1,22)):
    url = f"http://www.cnipa.gov.cn/jcdlzyqgg/index{n}.htm"

    response = requests.get(url)
    response.encoding = response.apparent_encoding
    soup = BeautifulSoup(response.text, 'lxml')
    for item in soup.find_all("li"):
        web[item.a.text] = item.a.attrs["href"]
```

```python
target_web = dict()
for i in web:
    if "专有权公告" in i:
        target_web[i] = web[i]
```

```python
owner = []
for page in target_web.keys():
    print("http://www.cnipa.gov.cn/jcdlzyqgg/" + target_web[page])
    url = "http://www.cnipa.gov.cn/jcdlzyqgg/" + target_web[page]
    response = requests.get(url)
    response.encoding = response.apparent_encoding
    owner.extend(re.findall("布图设计权利人：(.*?)<br />",response.text))
```

```python
pd.DataFrame(list(set(owner))).to_excel("集成电路布图权利人.xlsx",index=False)
```
